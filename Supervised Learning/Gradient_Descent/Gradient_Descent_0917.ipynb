```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X=np.array([1.0,1.5,2.5,4.0])
y=np.array([2.0,3.5,3.0,5.0])
M=len(y)

# Visualization of x,y
plt.figure(figsize=(12,8))
plt.scatter(X,y)
plt.show()
```

```python
# Loss function
def Loss(X,y,w,b):
    loss=sum(.5*(x*w+b-y[i])**2 for i,x in enumerate(X))/M
    return  loss

# partial derivative of w,b 
def gradL(X,y,w,b):
    partial_w=sum((x*w-y[i]*x)for i,x in enumerate(X))/M
    partial_b=sum((x*w-y[i])for i,x in enumerate(X))/M
    return partial_w,partial_b
    
```

```python
alpha=0.12 # Learning rate not too high or too small
# initial w,b
random_vec=np.random.rand(2)
w=random_vec[0]
b=random_vec[1]
# initial Loss
Loss(X,y,w,b)
```

```python
colors = {10:'blue', 20:'green', 30:'orange' , 50:'red'}
# Loss with new w,b
for i in range(51):
    dw,db = gradL(X, y, w, b) # start from initial partial w,b
    w -= alpha*dw
    b -= alpha*db
    L = Loss(X,y,w,b) # new Loss with new w,b
    if i in [10, 20, 30, 50]:
        print(f'Loss at iter{i} = {L}')
        # plot X, yhat with new w,b generated at i=10,20,30,50. **10 is the best with minimum loss
        plt.plot(X, [w*x+b for x in X], label = f'iter{i}', color = colors[i])
plt.legend()
plt.show()
```
